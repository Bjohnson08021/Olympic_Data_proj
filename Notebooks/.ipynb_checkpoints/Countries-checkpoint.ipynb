```python
import pandas as pd
```

```python
# Read in first file 
previous_olympics_df = pd.read_csv ("../Data_Sets/Summer Olympics Medals 1976-2008/Summer-Olympic-medals-1976-to-2008.csv",  encoding='latin-1')

previous_olympics_df.head()
```

```python
# Read in second file

teams_df = pd.read_excel ("../Data_Sets/Tokyo_2020/Teams.xlsx", engine = 'openpyxl')

teams_df.head()
```

```python
# Grab needed columns
countries_df = previous_olympics_df[['Country','Country_Code']]

countries_df = countries_df.drop_duplicates()
countries_df = countries_df.dropna()

countries_df
```

```python
# Rename column 
countries_df2 = teams_df[['NOC','Name']]
countries_df2.rename(columns={"NOC": "Country"}, inplace = True)

countries_df2 = countries_df2.drop_duplicates()

countries_df2
```

```python
# Combine dataframes

frames = [countries_df,
          countries_df2]
countries_concat = pd.concat(frames)

countries_concat = countries_concat.drop('Name', 1)

countries_concat

```

```python
# Remove duplicate columns and reset index
countries_concat = countries_concat.drop_duplicates()
countries_concat.reset_index(drop=True, inplace=True)

countries_concat
```

```python
# Copy country name column and drop Country_Code column
countries_concat['NOC'] = countries_concat['Country']
countries_concat.drop(columns=('Country_Code'), inplace=True)

countries_concat
```

```python
# Drop duplicates 

countries_concat = countries_concat.drop_duplicates()

countries_concat
```

```python
# Replace Country names in Country column

countries_concat["Country"].replace({"Hong Kong*": "Hong Kong", 
                                     "Hong Kong, China": "Hong Kong",
                                     "Netherlands Antilles*":"Netherlands", 
                                    "Puerto Rico*":"Puerto Rico",
                                    "Soviet Union":"Russia",
                                    "Unified team":"Russia",
                                    "United States of America":"United States",
                                    "People's Republic of China":"China",
                                    "Bermuda*":"Bermuda",
                                    "Virgin Islands*":"Virgin Islands",
                                    "Macedonia":"North Macedonia",
                                    "ROC": "Russia"}, inplace=True)
```
