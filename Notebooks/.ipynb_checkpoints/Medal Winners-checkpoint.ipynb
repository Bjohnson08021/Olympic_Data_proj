```python
import pandas as pd
```

# Getting Data from Olympics 2021

```python
tokyo_medals = pd.read_excel ("../Data_Sets/Tokyo_2020/Medals.xlsx").rename(columns={"Team/NOC":"Country", "Total":"Medal"})
tokyo_medals
```

```python
First_medals = tokyo_medals[['Country','Gold','Silver','Bronze', 'Medal']].copy()
First_medals.head()
```

```python
First_medals['Year'] = 2021
First_medals
```

```python
tokyo_medals = First_medals[['Year','Country', 'Gold','Silver','Bronze', 'Medal']]
tokyo_medals.reset_index(drop=True, inplace=True)
tokyo_medals
```

# Getting Data from Olympics 1976 - 2008

```python
previous_olympics_df = pd.read_csv ("../Data_Sets/Summer Olympics Medals 1976-2008/Summer-Olympic-medals-1976-to-2008.csv",  encoding='latin-1')

previous_olympics_df.head()
```

```python
#Dropping unnwanted columnns and rows

medal_by_years = previous_olympics_df[['Year', 'Country', 'Medal']].copy()
medal_by_years
```

```python
#Filtering Gold Medals throuhout the years

find_gold = medal_by_years.loc[medal_by_years["Medal"] == "Gold"]
find_gold
```

```python
#Creating Dataframe with just GOLD medals grouped by year annd country

Gold_medals = find_gold.groupby(["Year","Country"]).count().rename(columns={"Medal":"Gold"})
Gold_medals.head()
```

```python
#Creating Dataframe with just SILVER medals grouped by year annd country

find_silver = medal_by_years.loc[medal_by_years["Medal"] == "Silver"]
Silver_medals = find_silver.groupby(["Year","Country"]).count().rename(columns={"Medal":"Silver"})
Silver_medals
```

```python
#Creating Dataframe with just BRONZE medals grouped by year annd country

find_bronze = medal_by_years.loc[medal_by_years["Medal"] == "Bronze"]
Bronze_medals = find_bronze.groupby(["Year","Country"]).count().rename(columns={"Medal":"Bronze"})
Bronze_medals
```

# Combine Gold, Silver and Bronze Columns from History

```python
Medals_History = Gold_medals.join(Silver_medals).join(Bronze_medals)
Medals_History
```

```python
Medals_History.fillna(0,inplace=True)
Medals_History
```

```python
Medals_History['Medal']= Medals_History['Gold'] + Medals_History['Silver'] + Medals_History['Bronze']
Medals_History
```

```python
Medals_History.reset_index(inplace=True)
Medals_History
```

```python
#Concatenate two frames: History and Tokyo

frames = [Medals_History, tokyo_medals]
medal_winners = pd.concat(frames)
medal_winners
```

```python
#Convert float to Integer

medal_winners['Year'] = medal_winners['Year'].astype(int)
medal_winners['Silver'] = medal_winners['Silver'].astype(int)
medal_winners['Bronze'] = medal_winners['Bronze'].astype(int)
medal_winners['Medal'] = medal_winners['Medal'].astype(int)
medal_winners
```

```python
medal_winners.to_csv("../Output/Medal_Winners.csv")
```
