```python
import pandas as pd
import numpy as np
```

```python
previous_olympics_df = pd.read_csv ("../Data_Sets/Summer Olympics Medals 1976-2008/Summer-Olympic-medals-1976-to-2008.csv",  encoding='latin-1')
previous_olympics_df.head()
```

```python
#dataframe
#Athlete,year and event
athlete_event = previous_olympics_df[['Sport','Discipline', 'Event']]
athlete_event
```

```python
athlete_event.drop_duplicates()
athlete_event
```

```python
athlete_event.dropna(inplace=True)
```

```python
athlete_event
```

```python
athlete_event2 = pd.DataFrame({"Discipline":["Artistic Gymnastics","Karate","Wrestling","Rhythmic Gymnastics",
                                            "Baseball/Softball", "Artistic Swimming","Beach Volleyball", "Trampoline Gymnastics", "Marathon Swimming", "Canoe Slalom", "Surfing",
                                            "Canoe Sprint", "Cycling BMX Racing", "Rugby Sevens", "Equestrian", "Golf", "Skateboarding", "Modern Pentathlon", "Cycling Mountain Bike", "3x3 Basketball",
                                            "Cycling BMX Freestyle", "Sport Climbing"]})
```

```python
athlete_event = athlete_event.append(athlete_event2, ignore_index = True)

athlete_event.index = athlete_event.index + 1
```
