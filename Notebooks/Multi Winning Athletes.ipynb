```python
olympics_facts=pd.read_html("https://en.wikipedia.org/wiki/List_of_multiple_Olympic_gold_medalists")[0]
print(olympics_facts)
```

```python
olympics_facts.columns=["No.","Athlete","Nation", "Sport", "Years", "Games", "Gender", "Gold", "Silver", "Bronze", "Total"]
olympics_facts.set_index("No.",inplace=True)
olympics_facts
```

```python
olympics_facts.dropna(inplace=True)

```

```python
olympics_facts['Athlete'].replace({'Ahn Hyun-SooViktor Ahn':'Viktor Ahn'}, inplace=True)

olympics_facts['Nation'].replace({'South Korea Russia':'Russia',
                                 'East Germany Germany': 'Germany',
                                 'Russia ROC': 'Russia',
                                 'Unified Team Russia': 'Russia',
                                 'Unified Team Belarus': 'Belarus',
                                 'United Team of Germany West Germany':'Germany',
                                 'Soviet UnionÂ Unified Team':'Soviet Union',
                                 'Unified TeamÂ Russia':'Russia',
                                 'West GermanyÂ Germany':'Germany'}, inplace=True)


olympics_facts['Nation'] = olympics_facts['Nation'].str.replace(r'[^\x00-\x7f]', '')



olympics_facts['Sport'].replace({'Swimming, Water polo':'Swimming',
                                 'Water polo, Swimming': 'Water polo',
                                 'Gymnastics, Wrestling': 'Gymnastics',
                                'Table tennis': 'Table Tennis'}, inplace=True)

```

```python
olympics_facts['Gold'] = olympics_facts['Gold'].astype(int)
olympics_facts['Silver'] = olympics_facts['Silver'].astype(int)
olympics_facts['Bronze'] = olympics_facts['Bronze'].astype(int)
olympics_facts['Total'] = olympics_facts['Total'].astype(int)

olympics_facts
```
