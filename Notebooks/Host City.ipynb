```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

#engine = create_engine('sqlite://', echo=False)

#path = "~/datacode/Data_Olympics_proj_2/Data_Sets/Summer Olympics Medals 1976-2008/Summer-Olympic-medals-1976-to-2008.csv"
Olympic_big_df = pd.read_csv('../O', encoding ='latin1')

small_df = Olympic_big_df[['City', 'Year']]
small_df = small_df.dropna()

# small_df.to_sql('host_city', con=engine)
# engine.execute('SELECT * FROM host_city').fetchall()
small_df = small_df.drop_duplicates().reset_index()

small_df = small_df.drop(['index'], axis=1)
small_df
```

```python
new_data = [["London", 2012], ["Rio de Janeiro", 2016], ["Tokyo", 2021]]
new_df = pd.DataFrame(new_data, columns=['City', 'Year'])
new_small_df = small_df.append(new_df, ignore_index=True)
new_small_df['Year'] = new_small_df['Year'].astype(int)
new_small_df

new_small_df.index = new_small_df.index + 1

new_small_df
```

```python
new_small_df.to_csv('../Output/Host_City.csv')
```
