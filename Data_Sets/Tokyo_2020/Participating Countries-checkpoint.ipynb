```python
import pandas as pd
```

```python
previous_olympics_df = pd.read_csv ("../Data_Sets/Summer Olympics Medals 1976-2008/Summer-Olympic-medals-1976-to-2008.csv",  encoding='latin-1')

previous_olympics_df.head()
```

```python
teams_df = pd.read_excel ("../Data_Sets/Tokyo_2020/Teams.xlsx", engine = 'openpyxl')

teams_df.head()
```

```python
countries_df = previous_olympics_df[['Country','Country_Code']]

countries_df = countries_df.drop_duplicates()

countries_df
```

```python
countries_df2 = teams_df[['NOC','Name']]
countries_df2.rename(columns={"NOC": "Country"}, inplace = True)

countries_df2 = countries_df2.drop_duplicates()

countries_df2
```

```python
frames = [countries_df,
          countries_df2]
countries_concat = pd.concat(frames)

countries_concat = countries_concat.drop('Name', 1)

```

```python
countries_concat
```

```python
countries_concat = countries_concat.drop_duplicates()
```

```python
countries_concat.to_csv("../Output/Countries.csv")
```
